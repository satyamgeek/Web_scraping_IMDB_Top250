# Movie Info Scraping (Top 250 Imdb)

```python
import bs4
import lxml
import requests
```

```python
url = "https://www.imdb.com/chart/top/"
```

```python
result = requests.get(url)
```

```python
result.text
```

```python
soup = bs4.BeautifulSoup(result.text,'lxml')
```

```python
soup.find_all('title')
```

```python
movies = soup.select('td.titleColumn ')
```

```python
len(movies)
```

## Movie Title

```python
Movie_title = soup.select('td.titleColumn a')
```

```python
movie_list = [] #Extracting movie titles from the Movie_title
for t in Movie_title:
    Movie_name = t.getText()
    #print(Movie_name)
    movie_list.append(Movie_name)
print(movie_list)
```

## Movie Release

```python
release_date = soup.select('span .secondaryInfo')
```

```python
date_dic = []   #Extracting Year from the release_date
for date in release_date:
    Release_d = date.getText()[1:5]
    #print(Release_d)
    date_dic.append(Release_d)
print(date_dic)
```

## Movie Link

```python
site_list = []      #Extracting movie_link 
for link in soup.select('td.titleColumn a'):
    site="https://www.imdb.com"+str(link.attrs.get('href'))
    site_list.append(site)
print(site_list)
```

## Movie Director & Actors 

```python
Actor_list = []
Dir_list = []
for dirc in soup.select('td.titleColumn a'):     #Extracting movie director & actor names from the Movie_title
    direc = str(dirc.attrs.get('title'))
    director = direc.split('(dir.),')
    #print(director[0])
    #print(director[1])
    Dir_list.append(director[0])
    Actor_list.append(director[1])
print(Dir_list)
print(Actor_list)    
```

## Saving Data in CSV

```python
import csv
from csv import writer
```

```python
db = open("Movie_DB.csv", "w")  #Making a csv file to enter the collected Data
writer = writer(db)

for i in range(len(movies)):
    writer.writerow([movie_list[i], date_dic[i], Actor_list[i], Dir_list[i], site_list[i]])
    
db.close()
```

## Saving Data in SQLite

```python
import sqlite3

# Creating the DataBase and Creating the table
try:
    sqliteConnection = sqlite3.connect('SQLite_Movie.db')   
    sqlite_create_table_query = '''CREATE TABLE Movie_database1 (          
                                Movie_name TEXT PRIMARY KEY,
                                Release_date INTEGER NOT NULL,
                                Actors TEXT NOT NULL ,
                                Director TEXT NOT NULL,
                                Movie_link TEXT NOT NULL);'''

    cursor = sqliteConnection.cursor()
    print("Successfully Connected to SQLite")
    cursor.execute(sqlite_create_table_query)
    sqliteConnection.commit()
    print("SQLite table created")

    cursor.close()

except sqlite3.Error as error:
    print("Error while creating a sqlite table", error)
finally:
    if sqliteConnection:
        sqliteConnection.close()
        print("sqlite connection is closed")
```

```python
#Adding data to the Table

try:
    sqliteConnection = sqlite3.connect('SQLite_Movie.db')
    cursor = sqliteConnection.cursor()
    print("Successfully Connected to SQLite")
    for i in range(len(movies)):

        count = cursor.execute("""INSERT INTO Movie_database1
                              (Movie_name, Release_date, Actors, Director, Movie_link) 
                               VALUES (?,?,?,?,?)""",
                           (movie_list[i], date_dic[i], Actor_list[i], Dir_list[i], site_list[i])
                          )
    sqliteConnection.commit()
    print("Record inserted successfully into SqliteDb_developers table ", cursor.rowcount)
    cursor.close()

except sqlite3.Error as error:
    print("Failed to insert data into sqlite table", error)
finally:
    if sqliteConnection:
        sqliteConnection.close()
        print("The SQLite connection is closed")
```
